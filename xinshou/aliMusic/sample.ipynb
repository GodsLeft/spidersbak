```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
df = pd.read_csv('data/oneuservec.csv', index_col='ds', parse_dates=True)
df = df.iloc[:, 3:]
df = df.reset_index()
df = df.groupby('ds').sum()
df.plot()
```

```python
df = pd.read_csv('data/onesongvec.csv', index_col='ds', parse_dates=True)\
    .iloc[:, 3:]\
    .reset_index()\
    .groupby('ds').sum()
df.plot()
```

```python
df = pd.read_csv('data/oneartsvec.csv', index_col='ds', parse_dates=True)\
    .iloc[:, 3:]\
    .reset_index()\
    .groupby('ds').sum()
df.plot()
```

```python
df = pd.read_csv('data/oneartsallsong.csv', index_col=2, parse_dates=True, header=None)\
    .drop([0,1,4,5], axis=1)\
    .reset_index()\
    .groupby(2).mean()
```

```python
df.plot(style='ko--')
```

```python
df = pd.read_csv('data/oneartsallsong.csv', index_col=2, parse_dates=True, header=None)\
    .drop([1,4,5], axis=1)\
    .reset_index()\
    .groupby(2)\
    .agg({0:'count', 3:'mean'})
```

```python
fig, axes = plt.subplots(2,1)
df[3].plot(ax=axes[0], style='ko--')
df[0].plot(ax=axes[1], style='ko--')
```
